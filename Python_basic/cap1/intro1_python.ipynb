## 1.2 Operadores basicos de comparacion

```python
print(1 == 1)                            # =>  (igualdad)
print(1 != 2)                            # =>  (desigualdad)
print(1 > 0)                             # =>  (mayor que)
print(1 < 2)                             # =>  (menor que)
print(1 >= 1)                            # =>  (mayor o igual que)
print(1 <= 2)                            # =>  (menor o igual que)
```

### Comparaciones compuestas/combinadas

```python
print(1 < 2 < 3)                         # => (1 es menor que 2 y 2 es menor que 3)
print(1 < 3 > 2)                         # => (1 es menor que 3 y 3 es mayor que 2)
print(1 < 0 < 1/0)                       # => (Cortocircuito 1 no es menor que 0 por lo tanto todo lo que sigue es false)
print(1 == 1 == 1)                       # => (1 es igual a 1 y 1 es igual a 1)
print(1 == 1 > 0)                        # => (1 es igual a 1 y 1 es mayor que 0)
print(1 == 1 < 0)                        # => (1 es igual a 1 pero 1 no es menor que 0)
```

## 1.3 Operadores logicos en python


```python
#Valores booleanos (primitivos en python)
print(True)                               
print(False)       
```

### Operadores logicos nativos de python

```python
#Not 
print(not True)                         
print(not False)                           
```

```python
#And
print(True and True)                      
print(True and False)                     
print(False and True)                     
print(False and False)                    
```

```python
#Or
print(True or True)                       
print(True or False)                      
print(False or True)                      
print(False or False)                     
```

 Cortocircuito: En python, los operadores logicos and y or tienen un comportamiento de cortocircuito, lo que significa que no evaluan el segundo operando si el resultado ya se puede determinar con el primer operando.

```python
#Ejemplo de cortocircuito con el operador and
print(True and False and 1/0)             # =>  (no se evalua 1/0 porque el resultado ya es False)
print(True and True and 1/0)              # =>  (se evalua 1/0 porque el resultado es True)
print(False or True or 1/0)               # =>  (no se evalua 1/0 porque el resultado ya es True)
print(False or False or 1/0)              # =>  (se evalua 1/0 porque el resultado es False)
```

## 1.4 Strings "Cadena de caracteres"

```python
#Una String se define utilizando comillas simples '' o comillas dobles ""
print('Hola estoy aprendiendo python')       
print("Hola estoy aprendiendo python")
```

```python
#Si se utilizan """" se puede escribir una string multilinea
"""
Hola como
estas?
"""
```

```python
print("Hola" +  " mundo")                # => "concatenacion de strings
print("Hola " * 3)                       # => repeticion de strings
print("Hola" " mundo")                   # => Concanetacion automatica
print("Hola mundo"[5])                   # => acceso a un caracter por su indice (comienza en 0) Aqui la string funciona como una lista
```

```python
# Formateo de strings utilizando format

nombre = "Franco"
lugar = "La cancha"
horario = "a las 20:00"

"{} se junta con sus amigos en {}".format(nombre, lugar) 
"{0} se quedo dormido, {0} se levanto tarde, se tenia que juntar {2} en {1}" .format(nombre, lugar, horario) 
"El nombre es {nombre}, el lugar es {lugar} y el horario es {horario}".format(nombre=nombre, lugar=lugar, horario=horario) 
```

```python
#F-strings (formateo de strings utilizando f-strings)
f'{nombre} se junta con sus amigos en {lugar}' 
```

## 1.5 Objeto None


El objeto None representa en Python la ausencia de valor o un valor nulo. 
Es un tipo de dato especial que se utiliza para indicar que una variable no tiene ningún valor 
asignado o que una función no devuelve ningún resultado.


```python
Resultado = None
print(Resultado)
```

```python
#El none no representa ni un numero ni una string ni nada solo representa vacio
print(type(Resultado))
```

## 1.7 Conversion numerica en python

```python
# Decimal
print(str(10))
print(int("10"))
```

```python
# Binario
print(bin(10))    
print(int("0b1010", 2))
print(int("1010", 2))
```

```python
# Octal
print(oct(10))
print(int("0o12", 8))
print(int("12", 8))
```

```python
# Hexadecimal
print(hex(10))
print(int("0xa", 16))
print(int("0xA", 16))
print(int("a", 16))
print(int("A", 16))
```

## 1.8 Conversion de strings/numeros


```python
# En python no se utiliza el codigo ASCII como en C o Java, sino que se utiliza Unicode

# Se puede convertir una string a un numero y a su vez se puede pasar de un numero a una string

print(chr(89))                # Convierte el numero 89 a su caracter correspondiente en Unicode 
print(chr(560))

print(ord("Y"))              # Convierte el caracter "Y" a su numero correspondiente en Unicode
print(ord("Ȱ"))
```
