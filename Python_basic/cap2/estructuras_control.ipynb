## 2.1 Bloque If

```python
variable = 5
if variable >= 4:
    print("el valor es mayor a 4")
    
if variable:
    print("el valor es mayor a 0")
```

## 2.2 Bloque Else

```python
if variable >= 4:
    print("el valor es mayor a 4")
    
else:
    print("el valor es menor a 4")
```

### Caso de multiples condiciones

```python
if variable < 4:
    print("el valor es menor a 4")
    
elif variable > 4:
    print("el valor es mayor a 4")
    
else:
    print("el valor es mayor o igual a 4")
    
    
```

### Operador if in line


```python
# El if in line cumple la funcion de simplificar el mismo funcionamiento que un elif a partir de una sintaxis mas corta

edad = 25

mayor_edad = "Mayor de edad" if edad >= 18 else "Menor de edad"
print(mayor_edad)
```

## 2.3 Bucle for

```python
list(range(4))     # Crea una lista con los numeros del 0 al 3 [0, 1, 2, 3]

for i in range(4):
    print(i)        
```

```python
for animal in ["perro", "gato", "conejo"]:
    print(f"{animal} es un mamifero")
```

```python

nombres = ["Franco", "Sofia", "Borja", "Juan Ignacio"]

for indice, nombre in enumerate(nombres):     # La funcion enumerate agrega un indice a cada elemento de la lista, comenzando desde 0 por defecto
    print(f"{indice} - {nombre}")
    
```

```python
Edades = [18, 14, 17, 20]

for nombre, edad in zip(nombres, Edades):     # La funcion zip permite combinar dos listas en una sola, creando tuplas con los elementos correspondientes de cada lista 
    print(f"{nombre} tiene {edad} años")
```

```python
# Enumerate y zip pueden combinarse dentro del for
for indice, (nombre, edad) in enumerate(zip(nombres, Edades)):
    print(f"{indice} - {nombre} tiene {edad} años")
```

## 2.4 Bucle While

```python
# While
x = 0
while x < 5:    
    print(x)
    x += 1
```

```python
# Do while
x = 0
while True:
    print(x)
    x += 1
    if x >= 5:
        break
```

## 2.5 Estructura de control para errores

```python
# En python se utilizan try except y finally para manejar errores 
try: 
    resultado = 10 / 0
except ZeroDivisionError as exception:
   print(f"Ha ocurrido un error: {exception}")
   
```

```python
try:
    resultado = 10 / 0
except ZeroDivisionError as exception:
   print(f"Ha ocurrido un error: {exception}")
finally:
    print("Proceso finalizado")
```

### Errores comunes


```python
# IndexError
try:
    lista = [1, 2, 3]
    lista[5]
except IndexError as exception:
    print(f"Se ha utilizado indices fuera de rango: {exception}")
```

```python
# ZeroDivisionError
try:
    resultado = 10 / 0
except ZeroDivisionError as exception:
    print(f"No se puede dividir por cero: {exception}")
```

```python
#NameError
try:
    print(franco)
except NameError as exception:
    print(f"La variable no esta definida: {exception}")
```

```python
#ValorError
try:
    a,b = [1, 2, 3]
except ValueError as exception:
    print(f"Se han proporcionado mas valores de los esperados: {exception}")
    
```

```python
#KeyError
try:
    diccionario = {"a": 1, "b": 2}
    print(diccionario["c"])
except KeyError as exception:
    print(f"La clave no existe: {exception}")
```
