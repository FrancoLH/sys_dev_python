```python
def dividir(x: float, y: float) -> float:
    return x / y
```

```python
#Llamado de funcion

#El orden importa
print(dividir(10,2))
print(dividir(2,10))

#El orden no importa
print(dividir(x=10, y=2))
print(dividir(y=2, x=10))
```

```python
# return con multiples valores

precios: list[float] = [10.5, 20.0, 15.75, 15.0, 30.0, 25.5, 18.25, 22.0, 12.0, 28.5]

def oferta(precios: list[float]) -> tuple[bool, float]:
    precio_bajo = min(precios)
    if precio_bajo < 15.0:
        return True, precio_bajo
    return False, precio_bajo

print(oferta(precios))
hay_oferta, precio_oferta = oferta(precios)
print(hay_oferta, precio_oferta)
```

## 4.1 Parametros arbitrarios

```python
# Parametros posicionales para listas
def suma(*args: float):
    resultado = 0
    for valor in args:
        resultado += valor
        
    return resultado

suma(1, 2, 3, 4, 5)
```

```python
# Parametros de palabra clave para diccionarios

def imprimir_info(**kwargs: str):
    result = ""
    for arg in kwargs.values():
        result += arg + " "
    return result


imprimir_info(Nombre="Leonardo", Apellido="Heis", Edad="49", Profesion="Desarrollador")
```

## 4.2 Funcion lambda

```python
# La funcion lambda es una funcion anonima, es decir, no tiene nombre y se define en una sola linea

multiplicacion = lambda x, y: x * y

print(multiplicacion(5, 3))
```

```python
# Lo anterior es equivalente a:
def multiplicacion(x, y):
    return x * y

print(multiplicacion(5, 3))
```

### 4.2.1 Filter

```python
lista = [1, 2, 3, 4, 5, 6, 7, 8, 9]

def es_par(num):
    if num % 2 == 0:
        return True
    
numeros_pares = filter(es_par, lista)
print(list(numeros_pares))
    
```

```python
# Si utilizo la funcion lambda, puedo escribir lo anterior de la siguiente manera:
numeros_pares = filter(lambda num: num%2 ==0, lista)
print(list(numeros_pares))
```
