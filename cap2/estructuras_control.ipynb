## 2.1 Bloque If

```python
variable = 5
if variable >= 4:
    print("el valor es mayor a 4")
    
if variable:
    print("el valor es mayor a 0")
```

## 2.2 Bloque Else

```python
if variable >= 4:
    print("el valor es mayor a 4")
    
else:
    print("el valor es menor a 4")
```

### Caso de multiples condiciones

```python
if variable < 4:
    print("el valor es menor a 4")
    
elif variable > 4:
    print("el valor es mayor a 4")
    
else:
    print("el valor es mayor o igual a 4")
    
    
```

### Operador if in line


```python
# El if in line cumple la funcion de simplificar el mismo funcionamiento que un elif a partir de una sintaxis mas corta

edad = 25

mayor_edad = "Mayor de edad" if edad >= 18 else "Menor de edad"
print(mayor_edad)
```
