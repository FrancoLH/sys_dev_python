## 3.2 Listas en python

```python
lista = [] # lista vacia
print(lista)
```

```python
lista = [1, 2, 3] # Lista con valores asignados
print(lista)
```

```python
lista = [1, "Franco Heis", 3.14, True] # Lista con diferentes tipos de datos
print(lista)
```

#### Metodos de lista

```python
lista.append(10) # Agregar un elemento a la lista
print(lista)
```

```python
lista.extend([20, 30]) # Agregar varios elementos a la lista
print(lista)
```

```python
lista.pop() # Eliminar el ultimo elemento de la lista
print(lista)
```

```python
lista.insert(0, 0) # Insertar un elemento en una posicion especifica
print(lista)
```

```python
print(len(lista)) # Obtener la longitud de la lista
```
