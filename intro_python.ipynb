### Operadores basicos de comparacion

```python
print(1 == 1)                            # => True (igualdad)
print(1 != 2)                            # => True (desigualdad)
print(1 > 0)                             # => True (mayor que)
print(1 < 2)                             # => True (menor que)
print(1 >= 1)                            # => True (mayor o igual que)
print(1 <= 2)                            # => True (menor o igual que)
```

```python
from math import sqrt

print(sqrt(16))                            # => 4.0
```
